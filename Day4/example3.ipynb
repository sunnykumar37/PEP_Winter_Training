## SETS

```python
set_var = set()
print(set_var)
print(type(set_var))
```

```python
set_var = {1, 2, 3, 4, 5}
print(set_var)
```

```python
set_var =  {"Avenger", "Ironman", "SpiderMan"}
print(set_var)
type(set_var)
```

```python
set_var.add("Black Panther")
print(set_var)
```

```python
set1 = {"Avenger", "Ironman", "SpiderMan", "Hulk", "Thor"}
set2 = {"Avenger", "Thor", "Hulk", "Hitman"}
set2.intersection_update(set1)
print(set2)
```

```python
set1 = {"Avenger", "Ironman", "SpiderMan", "Hulk", "Thor"}
set2 = {"Avenger", "Thor", "Hulk", "Hitman"}
set2.difference(set1)
```
