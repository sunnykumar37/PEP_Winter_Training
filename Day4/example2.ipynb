### Replace the Char to another Char in a String

```python
a  = input("Enter the word: ")
b = input("Enter the character to be replaced: ")
c = input("Enter the symbol you want: ")
print(a.replace(b,c))
```

```python
n = input("Enter the word: ")
for char in n:
    if n.count(char)>1:
        print(char)
```

```python
a= input("Enter the string: ")
b = input("Enter the substring to be found: ")
c = input("Limit of string: ")
print(a.find(b, 0, int(c)))
```

```python
print(a.isalnum()) # check for alpha-numeric characters
print(a.isalpha()) # check for all char in the string are alphabets
print(a.isdigit()) # check for all char in the string are digits
print(a.istitle()) # check for first letter of each word is capital
print(a.isupper()) # check for all char in the string are uppercase
print(a.islower()) # check for all char in the string are lowercase
print(a.isspace()) # check for all char in the string are whitespace
print(a.startswith("s")) # check if string starts with "s"
print(a.endswith("y")) # check if string ends with "y"
```
