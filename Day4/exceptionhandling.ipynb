### Try Else

```python
try:
    a= int(input("Enter a number1: "))
    b= int(input("Enter a number2: "))
    c= a/b
    d= a*b
    e= a+b

except NameError:
    print("The user has not defined a variable")
except ZeroDivisionError:
    print("Please provide no greater than zero")
except TypeError:
    print("Try to make datatype similiar")
except Exception as e:
    print(e)
else:
    print(c)
    print(d)
    print(e)
```

```python
class DobException(Exception):
    pass
year = int(input("Enter the year of birth: "))
age= 2021 - year
try:
    if age <=30 & age>20:
        print("The age is valid. you can apply forthe exams")
    else:
        raise DobException
except DobException:
    print("The age is not within the range. you cannot apply for the exams")
```
