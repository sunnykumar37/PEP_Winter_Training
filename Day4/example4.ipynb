# Nested Dictionary

```python
car1={"Brand":"Ford","Model":"Mustang","Year":1964}
car2={"Brand":"BMW","Model":"X3","Year":2020}
car3={"Brand":"Audi","Model":"A6","Year":2019}
car_type ={"car1": car1, "car2": car2, "car3": car3}
print(car_type)
```

```python
print(car_type["car2"])
```

```python
print(car_type["car3"]["Year"])
```
