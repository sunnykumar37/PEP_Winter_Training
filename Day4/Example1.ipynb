```python
data =  {"Sunny": 21, "Mannat": 22, "Ankit": 20, "Rishabh": 20, "Arun": 19}
sorted_data = sorted(data.items(), key=lambda x: x[1], reverse=True)
for i in sorted_data:
    print(i[0], i[1])
```

```python
data = [{"name":  "Sunny", "age": 21},
        {"name":  "Mannat", "age": 22},
        {"name":  "Rishab", "age": 20},
        {"name":  "Arun", "age": 19}]
sorted_data = sorted(data, key=lambda x: x['age'], reverse=True)
for i in sorted_data:
    print(i['name'], i['age'])
```
