```python
class horse:
    sound  = 'neigh'
    movement = 'walks in attitude'

    def sound_horse(self):
        return self.sound
    
    def move_horse(self):
        return self.movement
    
def main():
        my_horse = horse()
        print(my_horse.sound_horse())
        print(my_horse.move_horse())
if __name__ == "__main__": 
    main()
```

```python
class Animals:
    def __init__(self, **kwargs):
        self.__type = kwargs['type'] if 'type' in kwargs else 'kitten'
        self.__name = kwargs['name'] if 'name' in kwargs else 'fluffy'
        self.__sound = kwargs['sound'] if 'sound' in kwargs else 'meow'

    def type(self, t = None):
        if t: self.__type = t
        return self.__type
    
    def name(self, n = None):
        if n: self.__name = n
        return  self.__name
    
    def sound(self, s = None):
        if s: self.__sound = s
        return self.__sound
    
    def __str__(self):
        return f'The {self.type()} is named {self.name()} and says {self.sound()}'
    
def main():
    a0 = Animals(type='kitten', name='fluffy', sound='meow')
    a1 = Animals(type='puppy', name='fido', sound='woof')
    print(a0)
    print(a1)
if __name__ == "__main__":
    main()
```

```python
class cars:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def car_info(self):
        return f"{self.make} was designed in {self.year} {self.model}"
    
def main():
        car0 = cars("Toyota", "Camry", 2020)
        print(car0.car_info())
        car1 = cars("BMW", "X5", 2021)
        print(car1.car_info())
        car3  = cars("Audi", "A4", 2019)
        print(car3.car_info())
    
if __name__ == "__main__":
    main()
```
