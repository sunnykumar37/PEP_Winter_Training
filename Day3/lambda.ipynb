# Lambda Function

```python
def add(a,b):
    return a+b
add(2,5)
```

```python
add2 =  lambda a,b : a+b
print(add2(2,5))
print(add(2,5) == add2(2,10))
```

### Lambda in sorting

```python
names = ['alice', 'bob', 'charlie', 'david', 'eve']
```

```python
sorted(names)
```

```python
sorted(names, key= lambda name: len(name))
```
