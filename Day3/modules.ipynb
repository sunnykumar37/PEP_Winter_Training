## Classes

##### Everythong is an object. You can define your own class to group common actions with data.

```python
class MyInt:
    '''Docstring for MyInt class'''
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return MyInt(self.value + other)
    
    def __repr__(self):
        return f"MyInt({self.value})"
```
